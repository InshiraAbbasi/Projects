```python
#The size has different values and it is not a number quantity.
#for that, we must convert the size column into bytes with a function that takes all 3 values and convert them to bytes.
def convert_size(size):
    if isinstance( size, str):
        if 'k' in size:
            return float(size.replace('k',"")) * 1024
        elif 'M' in size:
            return float(size.replace('M', "")) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
df['Size']=df['Size'].apply(convert_size)
```

```python
#checking changes
df['Size']
```

```python
# Renaming column to better understand we have converted size
df.rename(columns={'Size':'Size_in_Bytes'}, inplace=True) #inplace True is basically make the chnages True in your df
df.head()
```
