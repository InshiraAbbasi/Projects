The file star_data.csv contains a structured dataset with key information about various stars. Each row represents a unique star, and the columns provide the following details:

2. Temperature (K): The surface temperature of the star in Kelvin.
3. Luminosity (L/Lo): The luminosity of the star compared to that of the Sun.
4. Radius (R/Ro): The radius of the star relative to the Sun's radius.
5. Absolute Magnitude: The measure of the star's brightness as seen from a standard distance.
6. Distance (ly): The distance of the star from Earth in light-years.

## Lets categorize stars into temperature ranges.

```python
def classify_steller_on_temperature(temp):
    if 30000 <= temp <= 50000:
        return 'O'
    elif 10000 <= temp < 30000:
        return 'B'
    elif 7500 <= temp < 10000:
        return 'A'
    elif 6000 <= temp < 7500:
        return 'F'
    elif 5200 <= temp < 6000:
        return 'G'
    elif 3700 <= temp < 5200:
        return 'K'
    elif temp < 3700:
        return 'M'
    else:
        return 'Unknown'
df['Stellar Class']=df['Temperature (K)'].apply(classify_steller_on_temperature)
```

```python
df.head(10)
```

```python
#checking total numbers of each class
df['Stellar Class'].value_counts()
```

## Lets also give these classes a color according to the steller class standard.

```python
#The dictionary will give color to each class.
color_map = {
    'O': 'Blue',
    'B': 'Blue-White',
    'A': 'White',
    'F': 'Yellow-White',
    'G': 'Yellow',
    'K': 'Orange',
    'M': 'Red'
}

# Create a new column for color based on the Stellar Class
df['Color'] = df['Stellar Class'].map(color_map)
```

### We can see that the Stellar class is an object. Therefore, we'll need to encode this column inroder to predict as teh regression only uses numeric data.

```python
#importing library for Encoding data
from sklearn.preprocessing import LabelEncoder
```

```python
#Initializing the Model of encoder
label_encoder=LabelEncoder()
```

```python
#fit and transform the spectral class
df['Stellar Class Encoded']= label_encoder.fit_transform(df['Stellar Class'])
```

```python
#checking the chnages we have made
print(df[['Stellar Class', 'Stellar Class Encoded']].head())
```

```python
df.head()
```

### We have successfully encoded our column.

## We must Scale our columns so that the accuracy can be maintained. If the columns are used with such occuruing float values, there is a strong chance the accuracy might drop.
## Therefore, we must scale our Labels; Luminosity, Radius and Temperature.

```python
#Making our labels and features
X=df[['Luminosity (L/Lo)','Radius (R/Ro)','Temperature (K)']]
Y=df['Stellar Class Encoded']
```

```python
#calling a function to split, train and test our data
from sklearn.model_selection import train_test_split
```

```python
#splitting the data here
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.40)
```

```python
x_train
```

```python
y_train
```

```python
#Here, well do linear regression
from sklearn.linear_model import LinearRegression
```

```python
#making a vraible for linear regression and fitting training data in it.
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
prediction=reg.predict(x_test)
```

```python
plt.scatter(y_test, prediction)
```

### We will Evaluate the model for accuracy

```python
#importing libraries for validation
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
#They may give some reoccuring values so we'll round off
rounded_predictions=prediction.round()
```

```python
#evaluate accuracy
accuracy=accuracy_score(y_test, rounded_predictions)
print(f"Accuracy: {accuracy}")
```

```python
#making confusion matrix
cm=confusion_matrix(y_test,rounded_predictions)
print(f"Confusion Matrix: {cm}")
```
